```python
import pandas as pd
df = pd.read_excel('database/catalogue.xlsx')
for col in ['Products', 'Height (m)']:
    df[col] = df[col].fillna('Unspecified')
    
for col in ['Erupted volume', 'Anomaly volume']:
    df[col] = df[col].fillna(1)
```

```python
df.head()
```

```python
    import plotly.graph_objects as go
    key = 'Erupted volume'
    df['hover_text'] = df['Name'] + '<br>' + key + ': ' + (df[key]).astype(str) + '<br><br>References:<br>' + df['REF'] 
    scale = 1e-2
    marker = go.scattergeo.Marker(size=df[key].fillna(1)**(1/3)/scale,
                                  sizemode='area')
    data = go.Scattergeo(lon=df['Longitude'], lat=df['Latitude'],
                         marker=marker, text=df['hover_text'])
    layout = go.Layout(geo=go.layout.Geo())
    fig = go.Figure(data=data, layout=layout)
#     fig.show()
    
#     import plotly.io as pio
#     pio.write_html(fig, file='index.html', auto_open=True)
```

```python
import plotly.express as px
import plotly.io as pio

fig = px.scatter_geo(df, lon='Longitude', lat='Latitude', color='Products',
                     hover_name='Name',
                     hover_data=['Type', 'Anomaly volume'],  
                     size=df['Anomaly volume']**(1/3), size_max=50,
                     projection="natural earth",
                     width=1e3, height=800)
fig.show()    
pio.write_html(fig, file='index.html', auto_open=False)
```
